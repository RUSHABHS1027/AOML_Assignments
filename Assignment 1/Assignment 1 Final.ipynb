```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score
import pickle
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
df
```

```python
df["BMI_Category"] = np.where(df["BMI"] < 18.5, "Underweight", 
np.where(df["BMI"] < 24.9, "Healthy Range", 
np.where(df["BMI"] < 29.9, "Overweight", 
np.where(df["BMI"] < 39.9, "Obesity", 
np.where(df["BMI"] > 40, "Severe Obesity", "Unknown")))))
```

```python
df
```

```python
train, val = train_test_split(df, test_size=0.2, random_state=42)
```

```python
train
```

```python
val
```

```python
target = "Outcome"
numeric_features = ["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]
categorical_features = ["BMI_category"]
```

```python
X_train = train[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age","BMI_Category"]]
y_train = train["Outcome"]
X_val = val[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age", "BMI_Category"]]
y_val = val["Outcome"]
```

```python
scaler = StandardScaler()
X_train[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]] = scaler.fit_transform(X_train[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]])
X_val[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]] = scaler.transform(X_val[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]])
```

```python
X_train
```

```python
X_val
```

```python
encoder = OneHotEncoder(sparse_output=False, drop="first", handle_unknown="ignore")
X_train_encoded = pd.DataFrame(encoder.fit_transform(X_train[["BMI_Category"]]), 
                               columns=encoder.get_feature_names_out(["BMI_Category"]), 
                               index=X_train.index)
X_val_encoded = pd.DataFrame(encoder.transform(X_val[["BMI_Category"]]),
                             columns=encoder.get_feature_names_out(["BMI_Category"]), 
                             index=X_val.index)
```

```python
X_train = pd.concat([X_train[numeric_features], X_train_encoded], axis=1)
X_val = pd.concat([X_val[numeric_features], X_val_encoded], axis=1)
```

```python
best_k = None
best_knn_f1 = 0
for k in [3, 5, 7]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_val)
    f1 = f1_score(y_val, y_pred, average="micro")
    print(f"KNN (k={k}) F1 Score: {f1}")
    if f1 > best_knn_f1:
        best_k = k
        best_knn_f1 = f1
```

```python
best_depth = None
best_tree_f1 = 0
for depth in [3, 5, 7]:
    tree = DecisionTreeClassifier(max_depth=depth, random_state=42)
    tree.fit(X_train, y_train)
    y_pred = tree.predict(X_val)
    f1 = f1_score(y_val, y_pred, average="micro")
    print(f"Decision Tree (max_depth={depth}) F1 Score: {f1}")
    if f1 > best_tree_f1:
        best_depth = depth
        best_tree_f1 = f1
```

```python
#Take the features as input for a test sample

test_sample = pd.DataFrame({
    "Pregnancies": [6],
    "Glucose": [148],
    "BloodPressure": [72],
    "SkinThickness": [35],
    "Insulin": [0],
    "BMI": [33.6],
    "DiabetesPedigreeFunction": [0.627],
    "Age": [50],
    "BMI_Category": ["Obesity"]
})
```

```python
#Save the standard scaler and encoder
with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

with open("encoder.pkl", "wb") as f:
    pickle.dump(encoder, f)
```

```python
#Save the best model
best_model = KNeighborsClassifier(n_neighbors=best_k)
best_model.fit(X_train, y_train)
with open("model.pkl", "wb") as f:
    pickle.dump(best_model, f)
```

```python
#Load the above in the inference script and apply them on the test sample

with open("scaler.pkl", "rb") as f:
    scaler = pickle.load(f)

with open("encoder.pkl", "rb") as f:
    encoder = pickle.load(f)

with open("model.pkl", "rb") as f:
    model = pickle.load(f)

test_sample[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]] = scaler.transform(test_sample[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]])
test_sample_encoded = pd.DataFrame(encoder.transform(test_sample[["BMI_Category"]]),columns=encoder.get_feature_names_out(["BMI_Category"]),index=test_sample.index)
test_sample = pd.concat([test_sample[numeric_features], test_sample_encoded], axis=1)
```

```python
#Print the predicted Class
print(model.predict(test_sample))
```
